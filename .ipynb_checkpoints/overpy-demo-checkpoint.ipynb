```python
import overpy

API = overpy.Overpass()

# Fetch highways within Granollers.
# Working query (click to see results): http://overpass-turbo.eu/s/GR1

result = API.query("""
    area[name="Granollers"][admin_level=8];
    // query part for: “highway=*”
    //node["highway"]({{bbox}});
    (way["highway"](area);
    relation["highway"](area);
    );
    // print results
    out body;
    """)

```

```python
print(result)
```

Sin embargo, sí que hay contenido en la variable:

```python
for way in result.ways:
    print("Name: %s" % way.tags.get("name", "n/a"))
    print("  Highway: %s" % way.tags.get("highway", "n/a"))
    print("  Wikidata: %s" % way.tags.get("name:wikidata", "n/a"))
```
